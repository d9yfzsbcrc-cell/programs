```python
# Import the packages needed for the examples included in this notebook
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

Defino sistema continuo

```python
numc = [1]
denc = [1, 5]

sysc = ct.tf(numc, denc)
print(sysc)
```

Simulamos ante escalón

```python
salc = ct.step_response(sysc).plot()
```

Obtenemos el equivalente discreto

```python
T = 0.1
sysd = ct.sample_system(sysc, T, method='zoh')
print(sysd)
sald = ct.step_response(sysd).plot()
```

Defino un controlador proporcional

```python
k = 10

sal_serie = ct.series(sysd, k)
print(sal_serie)
sal_serie_k = ct.step_response(sal_serie).plot()
```

Lazo cerrado 

```python
sal_loop = ct.feedback(sysd, 1, sign = -1)
print(sal_loop)
sal_loop_res= ct.step_response(sal_loop).plot()
```
